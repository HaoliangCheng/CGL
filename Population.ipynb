1. Import important packages in python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

2. Import pupulation dataset

```python
df_population= pd.read_csv("https://data.wprdc.org/dataset/5b18c198-474c-4723-b735-cc5220ad43cc/resource/82f29015-6905-4b1c-8300-afe9bb2231b3/download/total-population.csv")
```

3. show dataset in table format

```python
df_population
```

```python
df_population.dtypes
```

4.  show dataset in graph format

```python
fig = plt.figure(figsize=(10, 6))
#plt.bar(df['EST'],df['Mean Humidity'], align = 'center') # doesn't guarantee proper ordering
plt.bar(range(len(df_population['Neighborhood'])),df_population['Estimate; Total'], align = 'center')
plt.xticks(range(len(df_population['Neighborhood'])), df_population['Neighborhood'],\
               rotation = 90, fontsize = 8)
plt.xlabel('neighborhood name')
plt.ylabel('population')
plt.title('population in different neighborhood in Pittsburgh')
plt.show()
```

5. Data Analysis

```python
data_sorted = df_population.sort_values(['Estimate; Total'], ascending=False)
data_sorted
```

```python
mean=df_population['Estimate; Total'].mean()
print("The average population in each neighborhood is "+str(mean))
median=df_population['Estimate; Total'].median()
print("The median population in each neighborhood is "+str(median))
min=df_population['Estimate; Total'].min()
print("The min population in each neighborhood is "+str(min)+". It is Chateau.")
max=df_population['Estimate; Total'].max()
print("The max population in each neighborhood is "+str(max)+". It is Squirrel Hill South.")
sum=df_population['Estimate; Total'].sum()
print("The sum population in Pittsburgh is "+str(sum))
```
