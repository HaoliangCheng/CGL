```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
firearms = pd.read_csv("firearms.csv", sep=",")
```

```python
firearms['neighborhood'].unique()
neighborhood_counts = firearms['neighborhood'].value_counts()
df = pd.DataFrame(neighborhood_counts)
new_column_names = ['Count']
df.columns = new_column_names
display(df)
```

```python
df.dtypes
```

```python
fig = plt.figure(figsize=(10, 6))
plt.xticks(range(len(df)), df.index, rotation=90, fontsize=7)
plt.bar(range(len(df)), df['Count'], align = 'center')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Arrests Involving Firearms')
plt.title('Arrests involving firearms in different neighborhood in Pittsburgh')
plt.show()
```
